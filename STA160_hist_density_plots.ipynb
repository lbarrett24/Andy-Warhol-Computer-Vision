```python
def plot_rgb_distribution(image):
    # Convert the image to RGB mode
    image = image.convert("RGB")

    # Convert the image to a NumPy array
    image_array = np.array(image)

    # Extract the R, G, and B channels
    red_channel = image_array[:, :, 0].flatten()
    green_channel = image_array[:, :, 1].flatten()
    blue_channel = image_array[:, :, 2].flatten()

    # Plotting the distribution of R, G, and B values
    plt.hist(red_channel, bins=256, color='red', alpha=0.5, label='Red', density = True)
    plt.hist(green_channel, bins=256, color='green', alpha=0.5, label='Green', density = True)
    plt.hist(blue_channel, bins=256, color='blue', alpha=0.5, label='Blue', density = True)

    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.legend()

    return plt
```

```python
orange_plot = plot_rgb_distribution(orange)
orange_plot.title('Distribution of R, G, B Values - Orange Image')
orange_plot.show()
```

```python
light_blue_plot = plot_rgb_distribution(light_blue)
light_blue_plot.title('Distribution of R, G, B Values -  Light Blue Image')
light_blue_plot.show()
```

```python
# Create a 2x3 grid of subplots
fig = plt.figure()

# orange_plot
ax1 = fig.add_subplot(2, 3, 1)
orange_plot = plot_rgb_distribution(orange)
orange_plot.title('Orange Image')

# light_blue_plot
ax2 = fig.add_subplot(2, 3, 2)
light_blue_plot = plot_rgb_distribution(light_blue)
light_blue_plot.title('Light Blue Image')

# red_plot
ax3 = fig.add_subplot(2, 3, 3)
red_plot = plot_rgb_distribution(red)
red_plot.title('Red Image')

# turquoise_plot
ax4 = fig.add_subplot(2, 3, 4)
turquoise_plot = plot_rgb_distribution(turquoise)
turquoise_plot.title('Turquoise Image')

# green_plot
ax4 = fig.add_subplot(2, 3, 5)
green_plot = plot_rgb_distribution(green)
green_plot.title('Green Image')

fig.tight_layout()
plt.legend()
plt.show()
```

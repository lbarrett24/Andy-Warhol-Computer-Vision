```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Load the images
light_blue = Image.open('/Users/lukebarrett/Desktop/UC Davis/STA Courses/STA 160/Data/light_blue.jpeg')
sage_blue = Image.open('/Users/lukebarrett/Desktop/UC Davis/STA Courses/STA 160/Data/sage_blue.jpeg')
red = Image.open('/Users/lukebarrett/Desktop/UC Davis/STA Courses/STA 160/Data/red.jpeg')
orange = Image.open('/Users/lukebarrett/Desktop/UC Davis/STA Courses/STA 160/Data/orange.jpeg')
turquoise = Image.open('/Users/lukebarrett/Desktop/UC Davis/STA Courses/STA 160/Data/turquoise.jpeg')
```

```python
# Convert images to numpy arrays
light_blue_np = np.array(light_blue)
sage_blue_np = np.array(sage_blue)
red_np = np.array(red)
orange_np = np.array(orange)
turquoise_np = np.array(turquoise)
```

```python
# Plot RGB values in scatter plots
fig, axs = plt.subplots(2, 3, figsize=(10,6))
axs[0,0].scatter(light_blue_np[:,:,0].ravel(), light_blue_np[:,:,1].ravel(), c=light_blue_np[:,:,2].ravel(), s=1)
axs[0,0].set_title('Light Blue')
axs[0,1].scatter(sage_blue_np[:,:,0].ravel(), sage_blue_np[:,:,1].ravel(), c=sage_blue_np[:,:,2].ravel(), s=1)
axs[0,1].set_title('Sage Blue')
axs[0,2].scatter(red_np[:,:,0].ravel(), red_np[:,:,1].ravel(), c=red_np[:,:,2].ravel(), s=1)
axs[0,2].set_title('Red')
axs[1,0].scatter(orange_np[:,:,0].ravel(), orange_np[:,:,1].ravel(), c=orange_np[:,:,2].ravel(), s=1)
axs[1,0].set_title('Orange')
axs[1,1].scatter(turquoise_np[:,:,0].ravel(), turquoise_np[:,:,1].ravel(), c=turquoise_np[:,:,2].ravel(), s=1)
axs[1,1].set_title('Turquoise')
plt.tight_layout()
plt.show()
```

```python
#scaling values to 0-1
light_blue_np = light_blue_np / 255.0
sage_blue_np = sage_blue_np / 255.0
red_np = red_np / 255.0
orange_np = orange_np / 255.0
turquoise_np = turquoise_np / 255.0
```

```python
# Plot each image in 3D
fig = plt.figure(figsize=(16, 6))

# Plot light blue image
ax = fig.add_subplot(151, projection='3d')
r, g, b = light_blue_np[:,:,0], light_blue_np[:,:,1], light_blue_np[:,:,2]
pixel_colors = light_blue_np.reshape((r.shape[0] * r.shape[1], 3))
ax.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
ax.set_title('Light Blue')

# Plot sage blue image
ax = fig.add_subplot(152, projection='3d')
r, g, b = sage_blue_np[:,:,0], sage_blue_np[:,:,1], sage_blue_np[:,:,2]
pixel_colors = sage_blue_np.reshape((r.shape[0] * r.shape[1], 3))
ax.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
ax.set_title('Sage Blue')

# Plot red image
ax = fig.add_subplot(153, projection='3d')
r, g, b = red_np[:,:,0], red_np[:,:,1], red_np[:,:,2]
pixel_colors = red_np.reshape((r.shape[0] * r.shape[1], 3))
ax.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
ax.set_title('Red')

# Plot orange image
ax = fig.add_subplot(154, projection='3d')
r, g, b = orange_np[:,:,0], orange_np[:,:,1], orange_np[:,:,2]
pixel_colors = orange_np.reshape((r.shape[0] * r.shape[1], 3))
ax.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
ax.set_title('Orange')

# Plot turquoise image
ax = fig.add_subplot(155, projection='3d')
r, g, b = turquoise_np[:,:,0], turquoise_np[:,:,1], turquoise_np[:,:,2]
pixel_colors = turquoise_np.reshape((r.shape[0] * r.shape[1], 3))
ax.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
ax.set_title('Turquoise')

# Set common axis labels
fig.text(0.5, 0.04, 'Red', ha='center')
fig.text(0.04, 0.5, 'Green', va='center', rotation='vertical')
fig.text(0.92, 0.5, 'Blue', va='center', rotation='vertical')

```

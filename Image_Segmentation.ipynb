```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
light_blue = Image.open('/Users/kaleemezatullah/Desktop/STA160/light_blue.jpeg')
sage_blue = Image.open('/Users/kaleemezatullah/Desktop/STA160/sage.jpeg')
red = Image.open('/Users/kaleemezatullah/Desktop/STA160/red.jpeg')
orange = Image.open('/Users/kaleemezatullah/Desktop/STA160/orange.jpeg')
turquoise = Image.open('/Users/kaleemezatullah/Desktop/STA160/turquoise.jpeg')
```

```python
# Convert images to numpy arrays
light_blue_np = np.array(light_blue)
sage_blue_np = np.array(sage_blue)
red_np = np.array(red)
orange_np = np.array(orange)
turquoise_np = np.array(turquoise)
```

```python
# Plot RGB values in scatter plots
fig, axs = plt.subplots(2, 3, figsize=(10,6))
axs[0,0].scatter(light_blue_np[:,:,0].ravel(), light_blue_np[:,:,1].ravel(), c=light_blue_np[:,:,2].ravel(), s=1)
axs[0,0].set_title('Light Blue')
axs[0,1].scatter(sage_blue_np[:,:,0].ravel(), sage_blue_np[:,:,1].ravel(), c=sage_blue_np[:,:,2].ravel(), s=1)
axs[0,1].set_title('Sage Blue')
axs[0,2].scatter(red_np[:,:,0].ravel(), red_np[:,:,1].ravel(), c=red_np[:,:,2].ravel(), s=1)
axs[0,2].set_title('Red')
axs[1,0].scatter(orange_np[:,:,0].ravel(), orange_np[:,:,1].ravel(), c=orange_np[:,:,2].ravel(), s=1)
axs[1,0].set_title('Orange')
axs[1,1].scatter(turquoise_np[:,:,0].ravel(), turquoise_np[:,:,1].ravel(), c=turquoise_np[:,:,2].ravel(), s=1)
axs[1,1].set_title('Turquoise')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of RGB values
ax.scatter(light_blue_np[:, :, 0].ravel(), light_blue_np[:, :, 1].ravel(), light_blue_np[:, :, 2].ravel(), c='lightblue', s=1, label='Light Blue')
ax.scatter(sage_blue_np[:, :, 0].ravel(), sage_blue_np[:, :, 1].ravel(), sage_blue_np[:, :, 2].ravel(), c='mediumaquamarine', s=1, label='Sage Blue')
ax.scatter(red_np[:, :, 0].ravel(), red_np[:, :, 1].ravel(), red_np[:, :, 2].ravel(), c='red', s=1, label='Red')
ax.scatter(orange_np[:, :, 0].ravel(), orange_np[:, :, 1].ravel(), orange_np[:, :, 2].ravel(), c='orange', s=1, label='Orange')
ax.scatter(turquoise_np[:, :, 0].ravel(), turquoise_np[:, :, 1].ravel(), turquoise_np[:, :, 2].ravel(), c='turquoise', s=1, label='Turquoise')

# Set labels for x, y, z axes
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')

# Set plot title and legend
ax.set_title('RGB Values')
ax.legend()

plt.show()
```
